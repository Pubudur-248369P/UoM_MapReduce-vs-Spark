##Year wise carrier delay from 2003-2010

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Iteration': [1, 2, 3, 4, 5],
    'Hadoop_Runtime': [11, 6, 0.6, 1.1, 0.7],
    'Spark_Runtime': [56, 28, 28, 28, 28]
}

df = pd.DataFrame(data)

# Plotting the bar graph
bar_width = 0.35
bar_positions_hadoop = df['Iteration']
bar_positions_spark = df['Iteration'] + bar_width

plt.bar(bar_positions_hadoop, df['Hadoop_Runtime'], label='Hadoop', width=bar_width, color='blue', edgecolor='black')
plt.bar(bar_positions_spark, df['Spark_Runtime'], label='Spark', width=bar_width, color='orange', edgecolor='black')

plt.xlabel('Iteration')
plt.ylabel('Running Time (seconds)')
plt.title('Hadoop vs Spark: Running Time vs Iteration')
plt.legend()
plt.grid(True)
plt.show()
```

## f. Comparison for all the Queries with sample data for an iteration

```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Time taken by query (sec)': ['CarrierDelay', 'NASDelay', 'WeatherDelay', 'LateAircraftDelay', 'SecurityDelay'],
    'HiveQL': [11, 5.8, 5.3, 5.7, 5.6],
    'Spark-QL': [56, 26, 26, 62, 30]
}

df = pd.DataFrame(data)



# Plotting the data as a table
plt.figure(figsize=(10, 4))
table = plt.table(cellText=df.values, colLabels=df.columns, cellLoc='center', loc='upper center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.5)


plt.axis('off')

plt.title('Comparison for all the Queries with sample data for an iteration')
plt.show()
```

## g. Average time taken by Map Reduce and Spark each query

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Query': ['CarrierDelay', 'NASDelay', 'WeatherDelay', 'LateAircraftDelay', 'SecurityDelay'],
    'Hadoop_Runtime': [19.4, 1.78, 3.78, 3.8, 1.4],
    'Spark_Runtime': [33.6, 27.2, 27.2, 36.0, 28.4]
}

df = pd.DataFrame(data)

# Increase figure size
plt.figure(figsize=(12, 8))

# Plotting the bar graph
bar_width = 0.35
bar_positions_hadoop = range(len(df['Query']))
bar_positions_spark = [pos + bar_width for pos in bar_positions_hadoop]

plt.bar(bar_positions_hadoop, df['Hadoop_Runtime'], label='HiveQL', width=bar_width, color='blue', edgecolor='black')
plt.bar(bar_positions_spark, df['Spark_Runtime'], label='Spark-SQL', width=bar_width, color='red', edgecolor='black')

plt.xlabel('Queries')
plt.ylabel('Average Time (seconds)')
plt.title('HiveQL vs Spark-SQL: Time Taken By')
plt.xticks([pos + bar_width / 2 for pos in bar_positions_hadoop], df['Query'])  # Centering the x-tick labels
plt.legend()
plt.grid(True)
plt.show()
```
